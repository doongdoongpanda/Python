```python
#while 반복문 연습
#다른유형의 while문 수행방법 1번째
i=0 #i에 0를 부여 
sum=0 #합계에 0을 부여 
while i <100 : #i가 0부터 100되기 전까지 반복한다. 
    i += 1 #i= i+1로서 1씩 증가한다고 생각하면 된다. 
    sum = sum + i #합계는 i값의 누적이다. 1+2+3+4+5..이런식으로 누적.

print("sum= ",sum) #while 반복을 빠져 나와서 출력한다.
```

```python
#다른유형의 while문 수행방법 2번째
i=0#i에 0를 부여 
sum=0#합계에 0을 부여 
while i <100 : #i가 0부터 100되기 전까지 반복한다.
    sum = sum + i #0,1,3,6,10,..4851, 4950 이 된다.
    i+=1
print("sum=",sum)
```

```python
#다른유형의 while문 수행방법 3번째
i=0#i에 0를 부여 
sum=0#합계에 0을 부여 
while i <101 :#i가 0부터 100되기 전까지 반복한다.
    sum = sum + i#0,1,3,6,10,..4851, 4950 이 된다.
    i+=1 #i= i+1로서 1씩 증가한다고 생각하면 된다. 100번째의 1값이 100
print("sum=",sum)#while 반복을 빠져 나와서 출력한다.
```

```python
i=0 #i에 0를 부여 
for a in ["1",2,"3",55]: #a는 컬렉션 ["1",2,"3",55]을 하나씩 가지고 for루프를 수행
    print("a[",i,"] 자료형:",type(a)," 값:",a) #컬렉션 값을 다 수행하면 반복블록을 탈출하게 된다.
    i+=1 # 1씩 증가한다.
```

```python
for i in range(1,10,1): #시작과 증가폭은 각각 생략할 수 있다
    print(i,",",end="") #변수i를 1부터 10전까지 9번 실행한 값을 출력한다.
```

```python
for i in range(10): #생략하게 되면 기본값: 시작 = 0, 증가폭 = 1이다. 
    print(i,",",end="") #10되기 전까지 출력한다.
```

```python
for i in range(10,0,-1): #10에서 시작하여 0까지 -1증가폭으로 출력한다.
    print(i,",",end="") #10,9,8....2,1 으로 출력
```

```python
for i in range(10):#반복문에서 탈출하려면 break를 쓰면 된다. 
    if(i==3): #만약 i ==3이라면
        break #해당 루프를 탈출 
    print(i,",",end="") #반복문이 도는 동안 해당 내용을 수행한다.
```

```python
for i in range(10):#해당문장 아래를 실행하지 않고 다시 루프를 실행할 경우 continue를 쓴다.
    if(i==3): #조건 만약 i가 3dlfkaus 
        continue #문장 아래를 실행하지않고 반복문의 처음으로 이동 후 계속 반복수행
    print(i,",",end="") #반복문이 도는 동안 해당 내용을 수행한다. i==3 제외
```

```python
i=0;sum=0 #i와 sum에 0를 부여 
while True: #문장으로 무한루프 실행 괄호 없이 그냥 True 로 작성(앞에 대문자)
    i+=1;sum+=i #i이 1씩 증가하고 sum이 1, 2,3,4,5...씩 더해져서 증가
    if(i==100): break #i가 100이 되면 반복문을 나온다.
print("i=",i,"sum=",sum) #i와 sum값을 출력한다.
```

```python
#별 30개씩 찍기, 30번째에는 줄을 바꿔야함.
for i in range(1,121): #range(0,120)이 아니고 range(1,121)을 한 이유는i%30과 연관이 있음
    print("*",end="") #별을 출력한다.
    if(i%30==0): print(end="\n")#만약 i가 30으로 나누어떨어지면 줄내림을 한다.
```

```python
#구구단예제 
for i in range(1,10): #반복문을 1부터 10까지 범위를 설정한다.i는 1씩 증가 
    for j in range(1,10):#반복문을 1부터 10까지 범위를 설정한다. j는 1씩 증가
        print( i,"*",j,"=",i*j) #i*j는 i*j값으로 출력된다.
    print("********************") #j반복문이 끝아면 라인을 출력한다.
```

```python
#거꾸로 트리 찍기 연습
for i in range(1,10): #1부터 10까지 반복한다. 
    for j in range(i,10): #i부터 10까지 반복한다. 
        print( "*",end="") #첨자를 출력한다. 
    print(end="\n") #두번째 반복문을 나오면 줄내림을 한다.
```

```python
i=1 #idp 1의 값을 준다. 
while True: #무한반복루프를 돈다. 
    j=i #j가 i랑 같다.
    while True: # 무한 반복루프를 돈다. 
        if(j==10): break #조건문내용이면 반복문에서 나온다. 
        print( "*",end="") #반복문동안 별을 출력한다. 
        j+=1 #j가 1씩 증가한다. 
    if(i==10): break #만약 i가 10이 되면 첫번째 반복문에서 나온다. 
    print(end="\n") #반복문에 있는동안 시행하게되는내용: 줄내림을 한다. 
    i+=1 #i가 1씩 증가한다.
```

```python
Mon=1 #달의 시작은 1월이니깐 초기값을 1로하여 변수에 부여한다. 
for LastDay in [31,28,31,30,31,30,31,31,30,31,30,31]: #리스트 안에 끝날짜들을 입력한다. 
    print(Mon,"월 :",end="") #월을 순서대로 출력한다.
    for Day in range(1,LastDay+1): #반복문을 수행한다. 
        print(Day,end=",") #변수를 출력하고 ,를 출력한다. 
    print(end="\n") #날짜 반복문을 나오면 줄내림을 한다.
    Mon+=1 #월이 1씩 증가한다.
```

```python
#구구단 출력하기
for i in range(2,10,3): #예제에 2단으로 시작, 그래서 2부터 10미만까지로 범위설정 2,5,8이니깐 3씩 증가
    for j in range(2,10): #뒤에 곱해줄 변수 2부터 10미만까지로 범위설정. 1씩 증가. 
        print(i,"*",j,"=",i*j,end="\t") #1열 출력내용
        print(i+1,"*",j,"=",i*j,end="\t") #2열 출력내용
        if(i+2!=10): print(i+2,"*",j,"=",i*j,end="\n") #3열 출력내용
        else: print(end="\n") # 줄내림
    print(end="\n") #두번째 반복문나오면 줄내림 실행
```

```python
#구구단 출력하기 2탄
for i in range(2,5): # 반복문을 연다. 세로줄에 맞게 2부터 5미만까지로 범위를 설정한다. 
    for j in range(2,10): # 반복문을 연다. 가로줄에 맞게 2부터 10미만까지로 범위를 설정한다. 
        print(i,"*",j,"=",i*j,end="\t") #1열 출력내용
        print(i+3,"*",j,"=",i*j,end="\t") #2열 출력내용
        if(i+6!=10): print(i+6,"*",j,"=",i*j,end="\n")#3열 출력내용
        else: print(end="\n") # 줄내림
    print(end="\n")#두번째 반복문나오면 줄내림 실행
```
